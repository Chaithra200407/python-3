#**Mileage Prediction - Regression Analysis**
**Source:**
This dataset was taken from the StatLib library which is maintained at Carnegie Mellon University. The dataset was used in the 19os America Statistical Association Exposition.
**Data Set Information:**
This dataset is a slightly modified version of the dataset provided in the StatLib library. In line with the use by Ross Quinlan (1993) in predicting the attribute "mpg", 8 of the original instances were removed because they had unknown values for the "mpg" attribute. The original dataset is available in the file "auto-mpg.data-original".
"The data concerns city-cycle fuel consumption in miles per gallon, to be predicted in terms of 3 multivalued discrete and 5 continuous attributes." (Quinlan, 1993)
**Attribute Information:**
1. mpg: continuous
2. cylinders: multi-valued discrete
3. displacement: continuous
4. horsepower: continuous
5. weight: continuous
6. acceleration: continuous
7. model year: multi-valued discrete
8. origin: multi-valued discrete
9. car name: string (unique for each instance)

# Define Target Variable y and Feature X


```python
import pandas as pd # Make sure pandas is imported

df.columns # Display the columns using the attribute directly


my_index = pd.Index(['mpg','cylinders','displacement', 'horsepower' , 'weight',
        'acceleration','model_year','origin','range'],
       dtype='object')
print(my_index)
```

```python
y =df['mpg']
```

```python
y.shape
```

```python
x = df[['displacement','horsepower','weight','acceleration']]
```

```python
x.shape
```

```python
x
```

# Scaling data

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

```python
pd.DataFrame(x).describe()
```

# Train Test split Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.7, random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

predict Test Data

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_percentage_error , mean_absolute_error,r2_score
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

# Polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
```

```python
x_train2 = poly.fit_transform(x_train)
```

```python
x_test2 = poly.fit_transform(x_test)
```

```python
lr.fit(x_train2,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly = lr.predict(x_test2)
```

# Model Accuracy

```python
from sklearn.metrics import mean_absolute_percentage_error , mean_absolute_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred_poly)
```

```python
mean_absolute_percentage_error(y_test,y_pred_poly)
```
